```python
import pandas as pd

datos = pd.read_csv('celsius.csv')
```

```python
datos.info()
```

```python
datos.head()
```

```python
import seaborn as sb

sb.scatterplot(data=datos, x="celsius",y="fahrenheit", hue="fahrenheit", palette= "coolwarm")
```

```python
# Caracteristicas (x), etiqueta (y)
x = datos["celsius"]
y = datos["fahrenheit"]
```

```python
x,y
```

```python
# x.values
# [-40, -10,   0,   8,  15,  22,  38]
# [[-40], [-10],   [0],   [8],  [15],  [22],  [38]])
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
# Entrenamiento

modelo.fit(x_procesada, y_procesada)
```

```python
celsius = 7900
prediccion = modelo.predict([[celsius]])
print(f"{celsius} grados celsius son {prediccion} grados fahrenheit")
```

```python
modelo.score(x_procesada, y_procesada)
```
