```python
import pandas as pd

datos = pd.read_csv('estatura_peso.csv')

datos.info()
```

```python
datos.head()
```

```python
import seaborn as sb

sb.scatterplot(datos,x='estatura',y='peso', hue='peso', palette='coolwarm')
```

```python
# Caracteristicas (x), etiqueta (y)

x=datos['estatura']
y=datos['peso']

x,y
```

```python
x.values
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)

x_procesada, y_procesada
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

# entranmiento
modelo.fit(x_procesada, y_procesada)
```

```python
estatura = 2.00
prediccion_peso = modelo.predict([[estatura]])
print(f"La estatura de {estatura} m tiene un peso de {prediccion_peso} Kg")

modelo.score(x_procesada, y_procesada)
```
